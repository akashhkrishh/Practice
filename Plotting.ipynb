```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(1,50,5)
y = 2*x+23
fig = plt.figure(figsize=(5,3))
plt.plot(x,y,linewidth=2.0,color='r',marker='o')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}
names = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10,5))
plt.bar(names,values)
plt.show()
```

```python
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}
names = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10,5))
plt.barh(names,values,color="orange")
plt.title("Bar Graph Demo")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

```python
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]
plt.pie(sizes, labels = label,autopct='%1.1f%%',shadow=True,explode = (0,0.1,0,0))
plt.title("Pie Chart Demo")
plt.show()
```

```python
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]
plt.stackplot(x,y,colors = 'b', alpha = 0.5)
plt.plot(x,y,color = 'g')
plt.grid(True)
plt.show()
```
